*****Naive Bayes classifier*****

```python
# Function to calculate mean and standard deviation for each class
def calculate_statistics(data):
    class_statistics = {}
    classes = np.unique(data[:, -1])
    
    for class_label in classes:
        class_data = data[data[:, -1] == class_label][:, :-1]
        class_statistics[class_label] = {
            'mean': np.mean(class_data, axis=0),
            'std': np.std(class_data, axis=0)
        }
    
    return class_statistics

def calculate_probability(x, mean, std):
    exponent = np.exp(-((x - mean) ** 2) / (2 * (std ** 2) + 1e-19))  # Adding a small constant
    return (1 / (np.sqrt(2 * np.pi) * (std + 1e-19))) * exponent  # Adding a small constant

# Function to predict class for a given sample
def predict_class(sample, class_statistics):
    probabilities = {}
    
    for class_label, stats in class_statistics.items():
        probabilities[class_label] = np.prod(calculate_probability(sample, stats['mean'], stats['std']))
        
    return max(probabilities, key=probabilities.get)

# Train the model and get class statistics
class_statistics = calculate_statistics(np.column_stack((X_train, y_train)))

# Make predictions on the test set
nbc_predictions = [predict_class(sample, class_statistics) for sample in X_test]

# Evaluate the model
correct_predictions = np.sum(nbc_predictions == y_test)
nbc_accuracy = correct_predictions / len(y_test)

print("y_test")
print(y_test)
print("Naive Bayes classifier Predictions")
print(nbc_predictions)
print(f"Naive Bayes classifier Accuracy: {nbc_accuracy * 100:.4f}%")
```

***Perceptron***

```python
class Perceptron:
    def __init__(self, learning_rate, epochs):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        # Initialize weights and bias
        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        for epoch in range(self.epochs):
            for i in range(X.shape[0]):
                # Calculate the predicted output
                predicted = np.dot(X[i], self.weights) + self.bias

                # Update weights and bias
                update = self.learning_rate * (y[i] - predicted)
                self.weights += update * X[i]
                self.bias += update

    def predict(self, X):
        # Calculate predicted outputs
        predictions = np.dot(X, self.weights) + self.bias
        # Apply step function to convert to binary predictions
        binary_predictions = np.where(predictions >= 0, 1, 0)
        return binary_predictions

# Initialize and train the perceptron model
perceptron_model = Perceptron(learning_rate=0.0001, epochs=5000) # 0.0001 5000
perceptron_model.fit(X_train, y_train)

# Make predictions on the test set
perceptron_predictions = perceptron_model.predict(X_test)

# Print predictions and accuracy
print("Perceptron Predictions:", perceptron_predictions)
perceptron_accuracy = np.mean(perceptron_predictions == y_test)
print(f"Perceptron Accuracy: {perceptron_accuracy * 100:.4f}%")
```

**Artificial Neural Network (ANN)**

```python
class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size, learning_rate, epochs):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.learning_rate = learning_rate
        self.epochs = epochs

        # Initialize weights and biases
        self.weights_input_hidden = np.random.randn(self.input_size, self.hidden_size)
        self.bias_hidden = np.zeros((1, self.hidden_size))
        self.weights_hidden_output = np.random.randn(self.hidden_size, self.output_size)
        self.bias_output = np.zeros((1, self.output_size))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x)) + 1e-19

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def forward(self, X):
        # Feedforward pass
        self.hidden_input = np.dot(X, self.weights_input_hidden) + self.bias_hidden
        self.hidden_output = self.sigmoid(self.hidden_input)
        self.output_input = np.dot(self.hidden_output, self.weights_hidden_output) + self.bias_output
        self.predictions = self.sigmoid(self.output_input)
        return self.predictions

    def backward(self, X, y):
        # Backpropagation
        output_error = y - self.predictions
        output_delta = output_error * self.sigmoid_derivative(self.predictions)

        hidden_error = output_delta.dot(self.weights_hidden_output.T)
        hidden_delta = hidden_error * self.sigmoid_derivative(self.hidden_output)

        # Update weights and biases
        self.weights_hidden_output += self.learning_rate * self.hidden_output.T.dot(output_delta)
        self.bias_output += self.learning_rate * np.sum(output_delta, axis=0, keepdims=True)
        self.weights_input_hidden += self.learning_rate * X.T.dot(hidden_delta)
        self.bias_hidden += self.learning_rate * np.sum(hidden_delta, axis=0, keepdims=True)

    def train(self, X, y):
        for epoch in range(self.epochs):
            # Forward and backward pass for each training example
            for i in range(len(X)):
                input_data = X[i:i+1]
                target = y[i:i+1]

                # Feedforward
                self.forward(input_data)

                # Backward
                self.backward(input_data, target)

            if epoch % 100 == 0:
                loss = np.mean(np.square(y - self.predictions))
                #print(f"Epoch {epoch}, Loss: {loss}")

    def predict(self, X):
        # Predict using the trained network
        return np.where(self.forward(X) >= 0.5, 1, 0)

    def accuracy(self, y_true, y_pred):
        correct = np.sum(y_true == y_pred)
        total = len(y_true)
        return correct / total + 1e-19


# Assuming X_train, y_train, X_test, y_test are your training and testing data
input_size = X_train.shape[1]
hidden_size = 6 #6
output_size = 1
learning_rate = 0.001
epochs = 1000

# Instantiate and train the neural network
nn = NeuralNetwork(input_size, hidden_size, output_size, learning_rate, epochs)
nn.train(X_train, y_train.reshape(-1, 1))

# Make predictions on the test set
ann_predictions = nn.predict(X_test)
ann_accuracy = nn.accuracy(y_test.reshape(-1, 1),ann_predictions)
print(ann_predictions)
print(f"ANN Accuracy: {ann_accuracy * 100:.4f}%")
```

**XGBoost**


```python
class DecisionTree:
    def __init__(self, max_depth=None):
        self.max_depth = max_depth
        self.tree = None
        self.fitted = False

    def mean_squared_error(self, y):
        return np.mean((y - np.mean(y))**2)

    def find_best_split(self, X, gradients):
        m, n = X.shape
        if m <= 1:
            return None, None, None  # No split if there is only one sample

        parent_gradient = np.sum(gradients)

        best_gradient_loss = float('inf')
        best_feature = None
        best_threshold = None

        for feature in range(n):
            thresholds = np.unique(X[:, feature])
            for threshold in thresholds:
                left_indices = X[:, feature] <= threshold
                right_indices = ~left_indices

                if np.sum(left_indices) == 0 or np.sum(right_indices) == 0:
                    continue

                left_gradient = np.sum(gradients[left_indices])
                right_gradient = np.sum(gradients[right_indices])

                weighted_gradient_loss = (left_gradient**2) / (np.sum(left_indices) + 1e-9) + \
                                         (right_gradient**2) / (np.sum(right_indices) + 1e-9)

                if weighted_gradient_loss < best_gradient_loss:
                    best_gradient_loss = weighted_gradient_loss
                    best_feature = feature
                    best_threshold = threshold

        return best_feature, best_threshold, best_gradient_loss

    def fit(self, X, gradients, depth=0):
        if depth == self.max_depth or self.mean_squared_error(gradients) == 0:
            return np.sum(gradients) / (np.sum(np.abs(gradients)) + 1e-9)

        best_feature, best_threshold, best_gradient_loss = self.find_best_split(X, gradients)

        if best_feature is None:
            return np.sum(gradients) / (np.sum(np.abs(gradients)) + 1e-9)

        left_indices = X[:, best_feature] <= best_threshold
        right_indices = ~left_indices

        left_subtree = self.fit(X[left_indices], gradients[left_indices], depth + 1)
        right_subtree = self.fit(X[right_indices], gradients[right_indices], depth + 1)

        self.tree = (best_feature, best_threshold, left_subtree, right_subtree)
        self.fitted = True
        return np.sum(gradients) / (np.sum(np.abs(gradients)) + 1e-9)

    def predict_instance(self, instance):
        if not self.fitted:
            raise ValueError("Decision tree not fitted.")
        
        def traverse_tree(node, instance):
            feature, threshold, left_subtree, right_subtree = node
            if instance[feature] <= threshold:
                if isinstance(left_subtree, tuple):
                    return traverse_tree(left_subtree, instance)
                else:
                    return left_subtree
            else:
                if isinstance(right_subtree, tuple):
                    return traverse_tree(right_subtree, instance)
                else:
                    return right_subtree

        return traverse_tree(self.tree, instance)

    def predict(self, X):
        if not self.fitted:
            raise ValueError("Decision tree not fitted.")
        return np.array([self.predict_instance(instance) for instance in X])


class XGBoost:
    def __init__(self, n_estimators=3000, learning_rate=0.001, max_depth=5):
        self.n_estimators = n_estimators
        self.learning_rate = learning_rate
        self.max_depth = max_depth
        self.trees = []

    def fit(self, X, y):
        # Initialize predictions and gradients
        predictions = np.zeros_like(y, dtype=np.float64)
        gradients = 2 * (y - predictions)

        for _ in range(self.n_estimators):
            # Fit a decision tree to the negative gradient
            tree = DecisionTree(max_depth=self.max_depth)
            tree.fit(X, gradients)

            # Calculate the tree predictions
            tree_predictions = self.learning_rate * tree.predict(X)

            # Update predictions and gradients
            predictions += tree_predictions
            gradients = 2 * (y - predictions)

            # Append the tree to the list
            self.trees.append(tree)

    def predict(self, X):
        # Initialize predictions
        predictions = np.zeros(X.shape[0], dtype=np.float64)

        # Make predictions using each tree
        for tree in self.trees:
            tree_predictions = self.learning_rate * tree.predict(X)
            predictions += tree_predictions

        # Apply the sigmoid function to get the final predictions
        final_predictions = self.sigmoid(predictions)
        # Convert to binary predictions (0 or 1)
        binary_predictions = np.round(final_predictions)

        return binary_predictions

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

# Instantiate and train the XGBoost model
xgb_model = XGBoost(n_estimators=10, learning_rate=0.001, max_depth=10)
xgb_model.fit(X_train, y_train)

# Make predictions on the test set
xgb_predictions = xgb_model.predict(X_test)
print("Predictions")
print(xgb_predictions)

# Calculate accuracy
xgb_accuracy = np.mean(xgb_predictions == y_test)
print(f"XGBoost Accuracy: {xgb_accuracy * 100:.4f}%")
```
